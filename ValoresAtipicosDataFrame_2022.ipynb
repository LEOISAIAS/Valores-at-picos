```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Carga desde un archivo .xlsx sin indice
data= pd.read_excel('Gastos y costos 20-23.xlsx', sheet_name='2022')
```

```python
#Verificamos información del DataFrame
data.info()
```

```python
#Corroboramos valores nulos
valores_nulos=data.isnull().sum()
valores_nulos
```

# **PROCEDIMIENTO PARA ELIMINAR VALORES NULOS EN DATAFRAME**

```python
#Reemplazamos valores nulos del dataframe con "bfill"
data1 =data.fillna(method="bfill") 
data1
```

```python
#Corroboramos valores nulos del dataframe
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores nulos del dataframe con "mean"
#Realizamos una copia del dataframe
data2=data.copy()
data2=data2.fillna(round(data.mean(),1))
data2
```

```python
#Corroboramos valores nulos del dataframe
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Realizamos diagrama de caja o bigote de cada columna del dataframe
fig = plt.figure(figsize =(15, 8))
data2.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show() #dibujamos el diagrama
```

# **PROCEDIMIENTO "DESVIACIÓN ESTÁNDAR" PARA ELIMINAR OUTLIERS EN DATAFRAME**

```python
#Método aplicando desviación estandar. Encuentro los valores extremos
y=data2
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Encontramos Ouliers del Dataframe
outliers= data2[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el DataFrame
data3= data2[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data3
```

```python
#Eliminamos columnas innecesarias
data4= data3.drop(['Fecha','Folio','UUID','RFC','Proveedor','TIPO GASTO','Descripción','MP','Otros','Tipo','Status','Poliza'], axis=1)
data4
```

```python
#Corroboramos valores nulos del dataframe4
valores_nulos=data4.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data_clean=data4.copy()
data_clean=data_clean.fillna(round(data4.mean(),1))
data_clean
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data_clean.isnull().sum()
valores_nulos
```

```python

fecha=data2['Fecha']
folio=data2['Folio']
UUID=data2['UUID']
RFC=data2['RFC']
prov=data2['Proveedor']
tip=data2['TIPO GASTO']
descr=data2['Descripción']
mp=data2['MP']
tipo=data2['Tipo']
status=data2['Status']
poliza=data2['Poliza']


Datos_limpios1 = pd.concat([fecha,folio,UUID,RFC,prov,tip,descr,mp,tipo,status,poliza, data_clean], axis=1)
Datos_limpios1
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=Datos_limpios1.isnull().sum()
valores_nulos
```

```python
#Convertir DataFrame a CSV
Datos_limpios1.to_csv("Datos_limpios_desvest_2022.csv")
#descargar archivo filtrado en csv
from google.colab import files 
  
  
files.download("Datos_limpios_desvest_2022.csv")
```

#**PROCEDIMIENTO "CUANTILES" PARA SUSTITUIR OUTLIERS EN DATAFRAME**

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=data4

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Encontramos Outliers del Dataframe
outliers_iqr= data4[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Obtenemos datos limpios del Dataframe
data3_iqr= data2[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data3_iqr
```

```python
#Eliminamos columnas innecesarias
data4_iqr= data3.drop(['Fecha','Folio','UUID','RFC','Proveedor','TIPO GASTO','Descripción','MP','Otros','Tipo','Status','Poliza'], axis=1)
data4_iqr
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data4_iqr.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data5_iqr=data4_iqr.copy()
data5_iqr=data5_iqr.fillna(round(data4_iqr.mean(),1))
data5_iqr
```

```python
# Unimos la columnas cualitativas con el dataframe

fecha=data2['Fecha']
folio=data2['Folio']
UUID=data2['UUID']
RFC=data2['RFC']
prov=data2['Proveedor']
tip=data2['TIPO GASTO']
descr=data2['Descripción']
mp=data2['MP']
tipo=data2['Tipo']
status=data2['Status']
poliza=data2['Poliza']


Datos_limpios2 = pd.concat([fecha,folio,UUID,RFC,prov,tip,descr,mp,tipo,status,poliza, data5_iqr], axis=1)
Datos_limpios2
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=Datos_limpios2.isnull().sum()
valores_nulos
```

```python
#Convertir DataFrame a CSV
Datos_limpios2.to_csv("Datos_limpios_iqr_2022.csv")
```
